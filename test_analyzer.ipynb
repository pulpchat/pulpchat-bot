```python
def strip_and_combine_args():
    list_1 = []
    list_2 = []
    all_args = []
    
    list_1 = df.iloc[:,3]
    list_2 = df.iloc[:,4]
    
    for line in list_1:
        all_args.append(line)

    for line in list_2:
        all_args.append(line)
    
    return pd.Series(all_args)
```

```python
import spacy
import numpy as np
nlp = spacy.load("en_core_web_sm")
```

```python
def get_random():
    random = np.random.randint(0, len(all_args))
    random_arg = all_args[random]
    return random_arg

def get_arg_info(random_arg):
    random_arg = random_arg
    random_tokenized = nlp(random_arg)
    
    random_arg_token_info = []
    
    for token in random_tokenized:
        text = token.text
        lemma = token.lemma_
        pos = token.pos_
        tag = token.tag_
        dep = token.dep_
        shape = token.shape_
        is_alpha = token.is_alpha
        is_stop = token.is_stop
        
        temp = {
            "Text": text,
            "Lemma": lemma,
            "POS": pos,
            "Tag": tag,
            "Dep": dep,
            "Shape": shape,
            "is_alpha": is_alpha,
            "is_stop": is_stop,
            }
        
        random_arg_token_info.append(temp)
    
    rand_df = pd.DataFrame(random_arg_token_info)
    rand_df = rand_df.transpose()
        
    print(random_arg)
    return rand_df

def get_arg_tokens():
    random_arg = get_random()
    arg_tokens = get_arg_info(random_arg)
    return arg_tokens

# https://thispointer.com/python-find-indexes-of-an-element-in-pandas-dataframe/
def getIndexes(dfObj, value):
    ''' Get index positions of value in dataframe i.e. dfObj.'''
    listOfPos = list()
    # Get bool dataframe with True at positions where the given value exists
    result = dfObj.isin([value])
    # Get list of columns that contains the value
    seriesObj = result.any()
    columnNames = list(seriesObj[seriesObj == True].index)
    # Iterate over list of columns and fetch the rows indexes where value exists
    for col in columnNames:
        rows = list(result[col][result[col] == True].index)
        for row in rows:
            listOfPos.append((row, col))
    # Return a list of tuples indicating the positions of value in the dataframe
    return listOfPos

def find_keywords(dfObj):
    keywords = [
    "be",
    "should",
    "good",
    "bad",
    "better"
    "worse",
    "best",
    "worst"    
]
    for keyword in keywords:
        getIndexes(dfObj, keyword)
```
